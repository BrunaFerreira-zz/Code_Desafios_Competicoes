```python
import numpy as np

#TENTAR
def lista_bordas(linha,coluna):
    vetor = []
    for i in range(0,linha):
        for j in range(0,coluna):
            if ((i == 0) or (i == (linha)-1)):
                x = i
                y = j
                vetor.append([x,y])
    for j in range(0,coluna):
        for i in range(0,linha):
            if ((j == 0) or (j == (coluna)-1)):
                x = i
                y = j
                if [x,y] not in vetor:
                    vetor.append([x,y])
    return vetor

def verifica_posicao (x,y,matriz):
    n = len(matriz)
    m = len(matriz[0])
    # Verificar se a posição gerada é válida, se não ultrapassa o limite do tabuleiro. 
    # Verificar também se a posição ainda não foi visitada
    if ((x >= 0) and (x < n) and (y >= 0) and (y < m) ):
        return True
    return False

def proxima_posicao (x,y,seta):
    n_x = 0
    n_y = 0
    if (seta == '>'):
        n_x = x
        n_y = y+1
    elif (seta == '<'):
        n_x = x
        n_y = y-1
    elif (seta == '^'):
        n_x = x-1
        n_y = y
    elif (seta == 'v'):
        n_x = x+1
        n_y = y
    return n_x,n_y
  
def Gruntz(matriz,v1,v2,k):
    if((matriz[v1][v2])=='x'):
        return True
    else:
        seta=matriz[v1][v2]
        nx,ny = proxima_posicao(v1,v2,seta)
        if (verifica_posicao(nx,ny,matriz)):
            if(Gruntz(matriz,nx,ny,k)):
                return True
        else:
            return False
    
def TENTAR(MATRIZ,N,M,K):
    if((N>2) and (M<100) and (0<=K<=100)):
        bordas = lista_bordas(N,M)
        for i in range(len(bordas)):
            x = bordas[i][0]
            y = bordas[i][1]
            if (Gruntz(MATRIZ,x,y,K)):
                return(True)
            else:
                return(False)
```

```python
def teste (arquivo):
    f = open(arquivo, 'r')
    l = []
    l = [line.split() for line in f]
    tamanho = len(l)
    matriz = np.empty(tamanho-1, dtype=object)
    for i in range (1,len(l)):
        x = list(str(l[i]))
        v = x[2: -2:]
        matriz = np.vstack([matriz, v])
    matriz = matriz[1:]
    
    entrada = l[0]
    N = int(entrada[0])
    M = int(entrada[1])
    K = int(entrada[2])
    
    print("Entradas: N =",N," M =",M," K =",K)
    print("Matriz:")
    print(matriz)
    if(TENTAR(matriz,N,M,K)==True):
        print("\nSaida: Sim")
    if(TENTAR(matriz,N,M,K)==False):
        print("\nSaida: Nao")
    
```
